# Visualizing the Dataset: Gaining Insights through Data Visualization

```python
# Unleashing the Power of Data Visualization: Exploring Electric Range Trends from 1997 to 2024

sns.histplot(x='Electric Range', data=df, kde=True)
plt.title("Electric Range Distribution (1997-2024)")
plt.xlabel("Electric Range")
plt.ylabel("Frequency")
plt.show()
```

```python
# Visualization of Base MSRP by Make (1998-2024)

plt.figure(figsize=(10, 5))
sns.barplot(x='Make', y='Base MSRP', data=df)
plt.title("Base MSRP by Make")
plt.xticks(rotation=90)
plt.xlabel("Make")
plt.ylabel("Base MSRP")
plt.show()
```

```python
# Top 10 Most Common Electric Vehicle Models (1997-2024)


top_most = df['Model'].value_counts().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=top_most.values, y=top_most.index, palette='viridis')
plt.xlabel('Number of Models')
plt.ylabel('Electric Vehicle Model')
plt.title('Top 10 Most Common Electric Vehicle Models (1997-2024)')
plt.show()

```

The bar plot highlights the top 10 most common electric vehicle models from 1997 to 2024. It reveals that Model S and Model 3 are the most prevalent models, indicating their popularity and widespread adoption in the electric vehicle market.

```python
# Electric Vehicle Companies Located in Legislative Districts of Washington State

plt.figure(figsize=(10, 6))
sns.boxplot(x='Make', y='Legislative District', data=df)
plt.xticks(rotation=90)
plt.xlabel('Electric Vehicle Company')
plt.ylabel('Legislative District')
plt.title('Electric Vehicle Companies Located in Legislative Districts of Washington State')
```

The box plot illustrates the distribution of electric vehicle companies across legislative districts in Washington State. It shows that multiple companies are located in various districts, indicating a diverse presence of electric vehicle manufacturers in the state.

```python
# Top Electric Vehicle Manufacturers (1997-2024)

plt.figure(figsize=(8, 8))
top_manufacturers = df['Make'].value_counts().head()
plt.pie(top_manufacturers, labels=top_manufacturers.index, autopct='%1.1f%%')
plt.title('Top Electric Vehicle Manufacturers (1997-2024)')
plt.show()
```

The pie chart shows the top electric vehicle manufacturers from 1997 to 2024. Tesla emerges as the leading manufacturer, followed by Nissan and Chevrolet, indicating their significant contribution to the electric vehicle market.

```python
# Visualization of Tesla Electric Vehicle Models (1997-2024)

Tesla = df[df['Make'] == 'TESLA']
model_counts = Tesla['Model'].value_counts().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='index', y='Model', data=model_counts)
plt.title("Electric Vehicle Models by Tesla (1997-2024)")
plt.xlabel("Model")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

It can be observed that Tesla has a diverse range of electric vehicle models, with Model S being the most prominent, followed by Model 3 and Model X. This indicates Tesla's strong presence and innovation in the electric vehicle market.

```python
# Electric Vehicle Type Distribution (1997-2024)


plt.figure(figsize=(8, 8))
vehicle_type_counts = df['Electric Vehicle Type'].value_counts()
plt.pie(vehicle_type_counts, labels=vehicle_type_counts.index, autopct='%1.1f%%')
plt.title('Electric Vehicle Type Distribution (1997-2024)')
plt.show()


```

The pie chart illustrates the distribution of electric vehicle types from 1997 to 2024, showing the proportion of each type. It indicates that Battery Electric Vehicles (BEV) and Plug-in Hybrid Electric Vehicles (PHEV) are the most common types in the dataset.

```python
# Electric Range Distribution by Company (1997-2024)

plt.figure(figsize=(12, 6))
sns.scatterplot(x='Make', y='Electric Range', data=df, alpha=0.5, color='blue')
plt.xticks(rotation=90)
plt.xlabel('Company')
plt.ylabel('Electric Range')
plt.title('Electric Range Distribution by Company (1997-2024)')
```

Conclusion: The scatter plot represents the distribution of electric range used by different companies from 1997 to 2024. Each data point represents a specific electric vehicle model, with the x-axis indicating the company and the y-axis representing the electric range. The plot provides an overview of the range variation across different manufacturers and allows for visual comparison of range characteristics between companies.

```python
# Model Year Distribution by Company (1997-2024)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Make', y='Model Year', data=df)
plt.xticks(rotation=90)
plt.xlabel('Company')
plt.ylabel('Model Year')
plt.title('Model Year Distribution by Company (1997-2024)')
```

The box plot illustrates the distribution of model years for electric vehicles made by different companies from 1997 to 2024. Each box represents the interquartile range (IQR) of model years for a specific company, with the median value shown as a line within the box. The plot provides a visual summary of the spread and central tendency of model years within each company, allowing for comparisons and observations on the timing of vehicle releases.

```python
# Geographic Distribution of Electric Vehicle Locations

sns.scatterplot(y='Longitude',x='latitude',data=df)

```

The scatter plot depicts the geographic distribution of electric vehicle locations based on latitude and longitude coordinates.



The plot illustrates the spatial spread of electric vehicle locations, providing insights into their geographical distribution. The plot can help identify clusters or patterns in electric vehicle presence across different regions.

```python
# Pairwise Relationships of Selected Variables


sns.pairplot(data=df[['Postal Code','Model Year','Electric Range','Base MSRP','Legislative District','DOL Vehicle ID','2020 Census Tract','Longitude','latitude']])

```

The pairplot provides a visual representation of the pairwise relationships between the selected variables in the dataset. From the plot, we can observe the scatter plots for each pair of variables, as well as the histograms along the diagonal. This allows us to identify any patterns or correlations between the variables.
